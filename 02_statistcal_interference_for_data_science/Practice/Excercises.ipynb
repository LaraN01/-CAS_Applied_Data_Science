### Task: Get some desciptive statistics from the poisson pmf with mean 1.7

```python
from scipy.stats import poisson
```

```python
fig, ax=plt.subplots(1,1)
```

```python
mu= 1.7
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
```

```python
print('mean= %1.2f Var= %1.2f  Std= % 1.2f Skewness= %1.2f Kurtosis= %1.2f'% (mean, var, var**0.5, skew, kurt))
```

Plot the mode and the median of the above distribution

## T Distribution

```python
from scipy.stats import t
```

```python
df=8
mean, var, skewn, kurt = t.stats(df, moments='mvsk')
```

```python
print('Mean= %1.2f Variance= %1.2f Standard Deviation= %1.2f Skewness= %1.2f Kurtosis= %1.2f'%(mean, var, var**0.5, skew, kurt ))
```

```python
df=2
x=np.linspace(t.ppf(0.01,df), t.ppf(0.99,df),100)
plt.plot(x,t.pdf(x,df), color='b', label= 'df=2')

df=10
x=np.linspace(t.ppf(0.01,df), t.ppf(0.99,df),100)
plt.plot(x,t.pdf(x,df), color='y', label='df=10')
plt.legend()
plt.show()
```

## Gamma distribution

```python
from scipy.stats import gamma
```

```python
a = 2
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')
```

```python
print('Mean= %1.2f Variance= %1.2f Standard Deviation= %1.2f Skewness= %1.2f Kurtosis= %1.2f'%(mean, var, var**0.5, skew, kurt ))
```

```python
a=2
x=np.linspace(gamma.ppf(0.01,a), gamma.ppf(0.99,a),100)
plt.plot(x, gamma.pdf(x,a), color='r', label=a)
a=4
x=np.linspace(gamma.ppf(0.01,a), gamma.ppf(0.99,a),100)
plt.plot(x, gamma.pdf(x,a), color='b', label=a)
a=8
x=np.linspace(gamma.ppf(0.01,a), gamma.ppf(0.99,a),100)
plt.plot(x, gamma.pdf(x,a), color='y', label=a)
plt.legend(title='Value of a')
plt.show()

#a correspond to k in gamma distribution wikipedia
```
